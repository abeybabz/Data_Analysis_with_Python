```python
i = 10421.78
q = 1
t = round(0,2)
while i < 3000000:
    r = int(i) * 0.03
    e = round(r,2)
    t += r
    
    print("\n",q,"current cap", round(i,2))
    print("\t difference from previous:", e)
    print("\t\t\t cumulative profit:", round(t,2))
    
    q += 1 
    if q == 365:
        break
    i += r
    e +=  i
    
```
