```python
import pandas as pd
pharma_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Training_set_begs.csv')
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Testing_set_begs.csv')
```

```python
pharma_data.head()
```

```python
test_data.head()
```

```python
print(pharma_data.shape)
print(test_data.shape)

Pharma_data = pharma_data.drop('Survived_1_year', axis=1)
print(Pharma_data.shape)
data = pd.merge(Pharma_data, test_data, how='outer')

print(data.shape)
data.head()
```

```python
data.isnull().sum()
```

```python
data.fillna(0, inplace=True)
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

patient_smoker = le.fit_transform(data['Patient_Smoker'])
data['Patient_Smoker'] = patient_smoker


patient_rural_urban = le.fit_transform(data['Patient_Rural_Urban'])
data['Patient_Rural_Urban'] = patient_rural_urban

patient_mental_condition = le.fit_transform(data['Patient_mental_condition'])
data['Patient_mental_condition'] = patient_mental_condition

data= data[['ID_Patient_Care_Situation','Diagnosed_Condition','Patient_ID','Patient_Age','Patient_Body_Mass_Index',
                'Patient_Smoker','Patient_Rural_Urban','Patient_mental_condition','A','B','C','D','E','F','Z', 'Number_of_prev_cond' ]]
data.head()
```

```python
X = data.iloc[0:23097,:]
y = pharma_data.Survived_1_year

```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.4027, random_state=47)


```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(X_train, y_train)
```

```python
X_test = data.loc[23098:,:]
y_pred = log_model.predict(X_test)
print(log_model.predict(X_test))
```

```python
X_test["prediction"] = y_pred
submission = X_test[["prediction"]]
submission.to_csv("Sample_Submission.csv", index = False)
```
