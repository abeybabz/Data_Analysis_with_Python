```python
deas as pd
import numpy as np

pharma_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Training_set_begs.csv')
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Testing_set_begs.csv')
```

```python
pharma_data.fillna(0, inplace=True)
```

```python
pharma_data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

patient_smoker = le.fit_transform(pharma_data['Patient_Smoker'])
pharma_data['Patient_Smoker'] = patient_smoker


patient_rural_urban = le.fit_transform(pharma_data['Patient_Rural_Urban'])
pharma_data['Patient_Rural_Urban'] = patient_rural_urban

patient_mental_condition = le.fit_transform(pharma_data['Patient_mental_condition'])
pharma_data['Patient_mental_condition'] = patient_mental_condition

pharma_data= pharma_data[['ID_Patient_Care_Situation','Diagnosed_Condition','Patient_ID','Patient_Age','Patient_Body_Mass_Index',
                'Patient_Smoker','Patient_Rural_Urban','Patient_mental_condition','A','B','C','D','E','F','Z', 'Number_of_prev_cond', 'Survived_1_year']]
pharma_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

patient_smoker = le.fit_transform(test_data['Patient_Smoker'])
test_data['Patient_Smoker'] = patient_smoker


patient_rural_urban = le.fit_transform(test_data['Patient_Rural_Urban'])
test_data['Patient_Rural_Urban'] = patient_rural_urban

patient_mental_condition = le.fit_transform(test_data['Patient_mental_condition'])
test_data['Patient_mental_condition'] = patient_mental_condition

test_data= test_data[['ID_Patient_Care_Situation','Diagnosed_Condition','Patient_ID','Patient_Age','Patient_Body_Mass_Index',
                'Patient_Smoker','Patient_Rural_Urban','Patient_mental_condition','A','B','C','D','E','F','Z', 'Number_of_prev_cond']]


pharma_data.head()
```

```python
X = pharma_data.drop('Survived_1_year', axis=1)
y = pharma_data.Survived_1_year
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.40276, random_state=47)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
y_pred =log_model.predict(test_data)
```

```python
len(y_pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score
print("F1 Score: ",f1_score(y_test, y_pred))
```

```python
X_test["prediction"] = y_pred
submission = X_test[["prediction"]]
submission.to_csv("Sample_Submission.csv", index = False)
```
